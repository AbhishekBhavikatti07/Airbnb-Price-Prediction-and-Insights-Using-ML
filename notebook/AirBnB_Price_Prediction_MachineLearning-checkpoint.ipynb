### Data Cleaning & Preprocessing
- Handling Missing Values
- Missing values were treated using appropriate strategies:

**<u>Outlier Treatment</u>**

```python
Airbnb_data = Airbnb_data.drop_duplicates()
```

```python
Airbnb_data.shape
```

From the above output we can say that the data set does not have any Duplicate values as the numbers of rows before after dupicate are same. 

```python
round(100*(Airbnb_data.isnull().sum() / len(Airbnb_data.index)),2)
```

```python
drop_col = ['id','name','description','thumbnail_url',
    'first_review','last_review','neighbourhood','host_has_profile_pic','zipcode']
Airbnb_data.drop(columns = drop_col, inplace = True)
```

```python
# Creating the rating field where the rating are 0, as they can be new posting 
Airbnb_data['missing_rating']=Airbnb_data['review_scores_rating'].isna().astype(int)

# now filling the missing values with median
Airbnb_data['review_scores_rating'].fillna(Airbnb_data['review_scores_rating'].median(),inplace=True)
Airbnb_data['bathrooms'].fillna(Airbnb_data['bathrooms'].median(),inplace=True)
Airbnb_data['bedrooms'].fillna(Airbnb_data['bedrooms'].median(),inplace=True)           
Airbnb_data['beds'].fillna(Airbnb_data['beds'].median(),inplace= True)
```

```python
# removing the % from host_response_rate
Airbnb_data['host_response_rate'] = (Airbnb_data['host_response_rate'].str.replace('%','')).astype(float)
Airbnb_data['host_response_rate'].fillna(Airbnb_data['host_response_rate'].median(),inplace=True)
```

```python

Airbnb_data['host_since'] = pd.to_datetime(Airbnb_data['host_since'])
Airbnb_data['host_exp_years']=round(((pd.Timestamp.today()-Airbnb_data['host_since']).dt.days / 365),2)
Airbnb_data['host_exp_years'].fillna(Airbnb_data['host_exp_years'].median(),inplace=True)
Airbnb_data.drop('host_since',axis=1,inplace=True)
```

### Converting the Date columns into the years of Exp Host

```python
Airbnb_data['host_exp_years'] = round(Airbnb_data['host_exp_years'] ,2)
```

```python
Airbnb_data.dropna(inplace=True)
Airbnb_data.isna().sum()
```

### Converting Text to Numeric 

```python
Airbnb_data['cleaning_fee'] = Airbnb_data['cleaning_fee'].astype(int)
binary_map = {'t':1, 'f':0}
Airbnb_data['host_identity_verified'] = Airbnb_data['host_identity_verified'].replace(binary_map).astype(int)
Airbnb_data['instant_bookable']=Airbnb_data['instant_bookable'].replace(binary_map).astype(int)
```

### Label Encoding 

```python
LE = LabelEncoder()
Airbnb_data['bed_type'] = LE.fit_transform(Airbnb_data['bed_type'])
```

```python
target = 'log_price'
high_card_cols = ['city','property_type','amenities','room_type', 'cancellation_policy']
encoder = TargetEncoder(cols=high_card_cols)
Airbnb_data[high_card_cols] = encoder.fit_transform(
    Airbnb_data[high_card_cols],
    Airbnb_data[target]
)
```

### Data Splitting Strategy

**The dataset was split into:**

- Training set: For model learning
- Test set: For unbiased performance evaluation

This ensures robust evaluation and helps prevent overfitting.

```python
x = Airbnb_data.drop('log_price',axis=1)
y = Airbnb_data['log_price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### XGBoost Model Selection and Training

In this project, XGBoost Regression was selected as the primary modeling approach due to its strong performance in structured, tabular datasets and its ability to capture complex, non-linear relationships between features and the target variable.

Airbnb pricing is influenced by multiple interacting factors such as property type, location, amenities, and host characteristics. XGBoost is well-suited for this problem as it:
Handles non-linear feature interactions effectively
- Is robust to outliers and noisy data
- Incorporates regularization to reduce overfitting
- Performs well with both numerical and encoded categorical features

<u>During model development:</u>

- The model was trained using the training dataset to learn underlying pricing patterns.
- Hyperparameters were tuned to optimize model performance and balance bias–variance trade-offs.
- Validation data was used to ensure the model generalizes well to unseen listings.
- Validation data was used during hyperparameter tuning to prevent overfitting and ensure that the model generalizes well to unseen Airbnb listings

The final XGBoost model demonstrated strong predictive capability and was therefore selected to predict Airbnb listing prices and derive pricing insights.

```python
xgb_model = XGBRegressor(
    n_estimators = 500,
    learning_rate = 0.05,
    max_depth = 6,
    random_state = 42
)
xgb_model.fit(X_train, Y_train)
```

```python
Y_pred = xgb_model.predict(X_test)
print("MAE:", mean_absolute_error(Y_test, Y_pred))
print("RMSE:", np.sqrt(mean_squared_error(Y_test, Y_pred)))
print("R²:", r2_score(Y_test, Y_pred))
```

### Model Evaluation and Performance Interpretation

The performance of the XGBoost regression model was evaluated on the test dataset using the following metrics:

- Mean Absolute Error (MAE): 0.1044
- Root Mean Squared Error (RMSE): 0.2112
- R² Score: 0.9154

**<u>Interpretation of Results</u>**

The MAE of 0.1044 indicates that, on average, the model’s predicted log-prices differ from the actual log-prices by approximately 0.10 units, demonstrating high prediction accuracy.

The RMSE of 0.2112 suggests that larger prediction errors are relatively limited, indicating consistent model performance across listings.

The R² score of 0.9154 shows that the model explains over 91% of the variance in Airbnb listing prices, reflecting strong explanatory power and an excellent fit to the data.

**Overall, these results confirm that the XGBoost model performs exceptionally well in capturing the key factors that influence Airbnb pricing and is capable of generating reliable and accurate price predictions for new listings.**
